# Q1. Explain GET and POST methods.
Ans.
GET: GET is a method used to request data from a specified resource. It sends data in the URL query string, visible to everyone in the URL bar, and has length limitations. It is commonly used for fetching data from a server.

POST: POST is a method used to submit data to be processed to a specified resource. It sends data in the request body, which is not visible in the URL, making it suitable for sensitive or large amounts of data. It has no size limitations imposed by the HTTP protocol.



# Q2. Why is request used in Flask?
Ans.In Flask, request is a global object that represents the HTTP request sent by the client to the server. It allows you to access data submitted in forms (via POST or GET), files uploaded, cookies, headers, and more. It's essential for handling incoming data and interacting with client requests within your Flask application.

# Q3. Why is redirect() used in Flask?
Ans.redirect() is a function in Flask used to redirect the client to a different URL. It's commonly used after handling a form submission to redirect the user to a different page or to another route within the application. This helps in maintaining clean URLs and separating concerns between different parts of the application.

# Q4. What are templates in Flask? Why is the render_template() function used?
Ans.Templates in Flask are HTML files with placeholders for dynamic data. They allow you to separate the presentation (HTML structure) from the logic (Python code). render_template() is a function in Flask that renders a template by replacing placeholders with actual values and returns the resulting HTML to the client. It simplifies generating HTML responses dynamically based on data provided by the Python code.
